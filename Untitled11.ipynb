```python
#Ahmed Ahmed Abdelhady
import cv2
from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator
import numpy as np

# Load the YOLOv8 model
model = YOLO('C:/Users/ahmod/OneDrive/Desktop/yolov8M_carModel/content/runs/detect/train3/weights/best.pt')


video_path = "C:/Users/ahmod/OneDrive/Desktop/yolov8M_carModel/content/test3.mp4"
cap = cv2.VideoCapture(video_path)


frame_num = 0

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        
        frame_num += 1

        results = model(frame)
        # to loop at each frame
        for r in results:
            boxes = r.boxes
            #to loop at each box in the frame
            for box in boxes:
                #extract the bounding box coordinates
                b = box.xyxy 
                #to extract the class of the bounding box 
                c = box.cls  
                
                #calculate the centroide
                cx = (b[0][0] + b[0][2]) / 2
                cy = (b[0][1] + b[0][3]) / 2
                   
                print(f'Frame number: {frame_num}, Bounding box : ({cx},{cy})')

        #anotate the objects
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)
        #cv2.imshow("YOLOv8 Inference", frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
